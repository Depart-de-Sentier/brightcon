```python
substitutions = {
    'United Kingdom': 'UK',
    'Deutschland': 'Germany',
    'The Netherlands': 'Netherlands',
    'España': 'Spain',
    'Ch': 'Switzerland',
    'Brasil': 'Brazil',
    'Usa': 'USA',
    'United States': 'USA',
    'Sverige': 'Sweden',
    'Uk': "UK",
    'Italia': 'Italy',
    'Países Baixos': 'Netherlands',
}
```

```python
data = [substitutions.get(x, x) for x in data]
```

```python
data = Counter(data).most_common()
```

```python
singles = ['Plus 1 each from:'] + sorted([x for x, y in data if y == 1])
more_names = [x for x, y in data if y > 1]
more_values = [y for x, y in data if y > 1]
```

```python
plt.figure(figsize=(16, 8))
plt.bar(more_names, more_values, color='None', edgecolor='k')
plt.gca().axis("off")
plt.plot(plt.xlim(), [0, 0], lw=2, color='k')

for x, y, z in zip(range(100), more_names, more_values):
    if z > 8:
        plt.text(x - 0.15, z - 0.15, y, rotation=90, va='top', size=12)
    else:
        plt.text(x - 0.15, z + 0.5, y, rotation=90, va='bottom', size=12)
        
for x in range(1, 11):
    plt.text(-1.5, x * 5, x * 5)

for x, y in zip(singles, range(20, 0, -1)):
    plt.text(19, 1.2 * y, x, size=12)

    
plt.tight_layout()
    
plt.savefig("participants.png", dpi=300)
```
